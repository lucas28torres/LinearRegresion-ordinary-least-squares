## Modelo de Regresión Lineal Simple (MRLS)

El anterior gráfico muestra una nube de puntos bidimensional con una recta de regresión lineal.

Predecir el valor de la variable Y a partir del comportamiento de la variable X, a nivel exacto es imposible, puesto que influyen diversos factores desconocidos que alteran el valor de cada variable.


Así pues, la recta real es $y_{i} = \alpha + \beta x_{i} + u_{i}\ \forall\alpha,\beta \in \mathbb{R}$,
donde $u_{i}$ representa el termino de perturbación aleatoria, el cual hace referencia a la infuencia de factores desconocidos, aleatorios y a otras variables no incluidas en el estudio.

La recta que se calculará será: $\widehat{y} = \ \widehat{\alpha} + \widehat{\beta}x$


### Mínimos Cuadrados Ordinarios (MCO)


Sean $\left( x_{i},y_{i} \right)\ldots\left( x_{n},y_{n} \right)$ variables cuantitativas, se propone un modelo de regresión lineal simple para tratar de entender como la variable X afecta a la variable Y.

Se busca la recta $\widehat{y} = \ \widehat{\alpha} + \widehat{\beta}x$ que mejor se ajusta al conjunto de datos.

Sea $e_{i} = y_{i} - \widehat{y_{i}}$, el error i-ésimo, hace falta encontrar una recta que minimice los residuos al cuadrado.

$$\Phi\left( \widehat{\alpha},\widehat{\beta} \right) = \ \sum_{i = 1}^{n}e_{i}^{2} = \ \sum_{i = 1}^{n}\left( y_{i} - {\widehat{y}}_{i} \right)^{2} = \sum_{i = 1}^{n}\left( y_{i} - \widehat{\alpha} - \widehat{\beta}x_{i} \right)^{2}$$

Se tiene que enontrar $\widehat{\alpha}\text{ y }\widehat{\beta}$, tal que
$\Phi\left( \widehat{\alpha},\widehat{\beta} \right)$ es mínima.

$$\frac{\partial\Phi}{\partial\widehat{\alpha}} = 0\ \frac{\partial\Phi}{\partial\widehat{\beta}} = 0$$


$$\frac{\partial\Phi}{\partial\widehat{\alpha}} = \sum_{i = 1}^{n}{2\left( y_{i} - \widehat{\alpha} - \widehat{\beta}x_{i} \right)\left( - 1 \right) = 0 \Longleftrightarrow \ \sum_{i = 1}^{n}{\left( y_{i} - \widehat{\alpha} - \widehat{\beta}x_{i} \right) = 0 \Longleftrightarrow \ \sum_{}^{}{y_{i} - n}}}\widehat{\alpha} - \widehat{\beta}\sum_{}^{}{x_{i} = 0} \Longleftrightarrow \frac{1}{n}\sum_{}^{}{y_{i} - \widehat{\alpha} - \widehat{\beta}}\frac{1}{n}\ \sum_{}^{}{x_{i} = 0} \Longleftrightarrow \overline{Y} - \widehat{\alpha} - \widehat{\beta}\overline{X} = 0 \Longrightarrow \widehat{\mathbf{\alpha}}\mathbf{=}\overline{\mathbf{Y}}\mathbf{-}\widehat{\mathbf{\beta}}\overline{\mathbf{X}}$$



$$\frac{\partial\Phi}{\partial\widehat{\beta}} = \sum_{i = 1}^{n}{2\left( y_{i} - \widehat{\alpha} - \widehat{\beta}x_{i} \right)\left( x_{i} \right) =}0 \Longleftrightarrow \sum_{}^{}{\left( y_{i} - \widehat{\alpha} - \widehat{\beta}x_{i} \right)\left( x_{i} \right) = 0}\  \Longleftrightarrow \sum_{}^{}y_{i}x_{i} - \widehat{\alpha}\sum_{}^{}x_{i} - \widehat{\beta}\sum_{}^{}{x_{i}^{2} = 0} \Longleftrightarrow \frac{1}{n}\sum_{}^{}{y_{i}x_{i} -}\widehat{\alpha}\overline{X} - \widehat{\beta}\ \frac{1}{n}\sum_{}^{}{x_{i}^{2} = 0} \Longleftrightarrow \frac{1}{n}\sum_{}^{}{x_{i}y_{i}} - \left( \overline{Y} - \widehat{\beta}\overline{X} \right)\overline{X} - \widehat{\beta}\frac{1}{n}\sum_{}^{}{x_{i}^{2} = 0}\  \Longleftrightarrow \frac{1}{n}\sum_{}^{}{(x_{i}y_{i}} - \overline{X}\overline{Y}) - \ \widehat{\beta}(\frac{1}{n}\sum_{}^{}x_{i}^{2} - {\overline{X}}^{2}) = 0 
\Longleftrightarrow S_{\text{xy}} - \widehat{\beta}S_{x}^{2}\ \  \Longrightarrow \widehat{\beta} = \frac{S_{\text{xy}}}{S_{x}^{2}}$$

### Método de los Mínimos Cuadrados 

Como ejemplo elemental en la aplicación de éste criterio, revisaremos la
noción de media aritmética.

Consideramos una serie de datos $x_{1},x_{2}\ldots x_{n}$ y los
interpretamos como los resultados de $n$ medidas de una variable $X$.
Cualquier cantidad $x$ puede ser un valor de centralización de los
datos, siempre que cumpla alguna relación de proximidad. Si tomamos como
medida de desviación de la expresión,

$\delta\left( x \right) = \left( x - x_{1} \right)^{2} + \left( x - x_{2} \right)^{2} + \cdot \cdot \cdot + \left( x - x_{n} \right)^{2}$
obtendremos el valor x que minimiza esta función derivando e igualando a
0,
$\delta^{'}\left( x \right) = 2\left( x - x_{1} \right) + 2\left( x - x_{2} \right) + \cdot \cdot \cdot + 2\left( x - x_{n} \right) = 0$.
La solución única de esta ecuación es
$x = \overline{x} = \frac{x_{1} + x_{2} + \cdot \cdot \cdot x_{n}}{n}$.

Por lo tanto, la media aritmética es el valor de centralización que
minimiza la suma de los cuadrados de las desviaciones de los datos.

Se puede llegar al mismo resultado aplicando la teoría de la proyección
ortogonal y proyectando el vector $x = (x_{1},x_{2},\ldots,x_{n})$ sobre
la recta generada por el vecor $w = (1,1\ldots,1)$ el resultado es el
vector $\overline{x}\text{w.}$

Matricialmente $CΧ = d,\ C\mathcal{\in M(}\mathbb{R)}$.

En caso de ser incompatible, el criterio considera que como solución
aproximada se considerará aquella que minimice
$\left\| \ CΧ - d \right\|^{}$.

Si $c_{1},\ldots,c_{n}$ son las columnes de la matriz $C$, el sistema es
equivalente a
$x_{1}c_{1} + x_{2}c_{2} + \cdot \cdot \cdot + x_{n}c_{n} = d$.

La incompatibilidad equivale a decir que
$d\  \notin H = \left\langle (c_{1},c_{2}\ldots c_{n}) \right\rangle$,
por lo que hay que encontrar la proyección ortogonal de $d'$ de $d$
sobre $H$ y resolver
$x_{1}c_{1} + x_{2}c_{2} + \cdot \cdot \cdot + x_{n}c_{n} = d'$.

Se definen:

-   $CΧ = d^{'}\  \Longleftrightarrow AΧ = b$, donde $A = C^{T} \cdot C$
    y $b = C^{T} \cdot d$

-   $Χ$ es único $\Longleftrightarrow Rg\left( C \right) = n$ y
    $\det\left( A \right) \neq 0$

Demostraciones:

-   $CΧ = d^{'} \Longleftrightarrow \ CΧ - d \in H^{\bot}$, por lo que
    el producto escalar de cada generador de $H$ debe ser 0.

> $C^{T}\left( CΧ - d \right) = 0\  \Longrightarrow C^{T}CX = C^{T}d = 0 \Longrightarrow AΧ = b$

-   $\text{Rg}\left( c \right) = n \Longleftrightarrow CΧ = d^{'}\ $sea
    sistema compatible
    determinado$\Longleftrightarrow \det\left( A \right) \neq 0$
    $\blacksquare$

El sistema  a resolver es: $AΧ = b$ $\Leftrightarrow  (X^{T}X)^{-1}X^{T}b$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Uso del data set Boston.
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
print(boston.DESCR)
```

Fórmula de minimizar el error cuadrático medio (MCO): $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
X = np.array(boston.data[:,5])
Y = np.array(boston.target)

plt.scatter(X,Y, alpha = 0.3)

#Añadimos una columna de 1 para término independiente

X = np.array([np.ones(506), X]).T

#Multiplicación matricial @

B = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.plot([4,9],[B[0]+B[1] *4,B[0]+B[1]*9], c= "red")
plt.show()
```
